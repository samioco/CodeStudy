```python
import folium
import pandas
```

```python
data = pandas.read_csv("Volcanoes.txt")
map = folium.Map(location=[40, -115], zoom_start=4.5, tiles = "Stamen Terrain")
fg = folium.FeatureGroup(name="My Map")
```

```python
coords = [[data["LAT"][i],data["LON"][i]] for i in range(len(data)) if data["LAT"][i]]
```

```python
#initiate lists of individual volcano info columns
lat = list(data["LAT"])
lon = list(data["LON"])
elev = list(data["ELEV"])
```

```python
#Use coord list of 2 columns (lat/lon)
for c in coords:
    fg.add_child(folium.Marker(location=c, popup="Volcano", icon=folium.Icon(color='green')))
map.add_child(fg)
map.save("Map4a.html")
```

```python
#Use 2 separate lists of lat/lon
for y,x in zip(lat,lon):
    fg.add_child(folium.Marker(location=[y,x], popup="Volcano", icon=folium.Icon(color='green')))
map.add_child(fg)
map.save("Map4b.html")
```
