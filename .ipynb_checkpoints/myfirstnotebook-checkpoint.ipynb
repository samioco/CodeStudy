```python
s = 'How many times does each word show up in this sentence word word shoe up up show'
```

```python
words = s.split()
```

```python
words
```

```python
Counter(words)
```

```python
c = Counter(words)
        
```

```python
c
```

```python
c.most_common(2)
```

```python
c.most_common(3)
```

```python
sum(c.values())
```

```python
list(c)
```

```python
c.items()
```

```python
c.most_common()
```

```python
c.most_common()[:-2:-1]
```
