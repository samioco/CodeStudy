```python
#Dictionary Comprehensions
#Just like List Comprehensions, Dictionary Data Types also 
#support their own version of comprehension for quick creation. 
#It is not as commonly used as List Comprehensions, 
#but the syntax is:
```

```python
{x:x**2 for x in range(10)}
```

```python
#One of the reasons it is not as common is the difficulty in 
#structuring key names that are not based off the values.
```

```python
#Iteration over keys, values, and items
#Dictionaries can be iterated over using the keys(), values() 
#and items() methods. For example:
```

```python
d = {'k1':1,'k2':2}
```

```python
d
```

```python
for k in d.keys():
    print(k)
```

```python
for v in d.values():
    print(v)
```

```python
for item in d.items():
    print(item)
```

```python
#Viewing keys, values and items
#By themselves the keys(), values() and items() methods return a 
#dictionary view object. This is not a separate list of items. 
#Instead, the view is always tied to the original dictionary.
```

```python
key_view = d.keys()
key_view
```

```python
d['k3'] = 3
d
```

```python
key_view
```
